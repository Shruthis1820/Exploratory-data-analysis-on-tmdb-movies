# 2)What is the average number of words in movie titles between the year 2000-2005?

```python
# make a dataframe containing the release_year from '2000 to 2005' from the existing dataframe df
df_year_range = df[df['release_year'].isin([2000,2001,2002,2003,2004,2005])]
```

```python
# extract only the 'original_title' column from df_year_range
df_titles_cols = df_year_range[['original_title']]
df_titles_cols
```

```python
#convert the dataframe into a Series containing the 'original_title' for easy operation on data
list_titles = pd.Series((df_titles_cols['original_title']))
list_titles
```

```python
# split the strings of list_title into substring using str.split() function 
# splitting default by whitespace
title_split= list_titles.str.split()
title_split
```

```python
# count the number of substrings in each individual rows using str.len() function
title_count =title_split.str.len()
title_count
```

```python
title_count.value_counts()
```

```python
# computing the average of all the count values using mean()
title_count.mean()
```

# Observation:
The average number of words in movie titles between the year 2000-2005 is nearly equal to 3

# 6) Which Movie Has The Highest and Lowest Profit?

```python
df_three_cols=df[['revenue','budget','original_title']]
df_three_cols
```

```python
# profit is determined by subtracting budget from revenue
df['profit']= df_three_cols['revenue']- df_three_cols['budget']
```

```python
df.sort_values(by='profit', ascending=False).head(1)
```

```python
df.sort_values(by='profit', ascending=False).tail(1)
```

# Observation
From the data analysis, 'Avatar' has the highest profit of 2.544505847 billion dollars and 'The Warrior's Way' has a loss of 413.912431  million dollars.

# 7) Which Year Has The Highest Profit Rate?

```python
#year-wise analysis of profit earn by movies 
df.groupby('release_year')['profit'].mean().plot(xticks=np.arange(1960,2015,5))

#setup the title and labels of the figure.
plt.title("Year Vs Average Profit",fontsize = 14)
plt.xlabel('Release year',fontsize = 13)
plt.ylabel('Average Profit',fontsize = 13)

import seaborn as sns

#setup the figure size.
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style("whitegrid")
```

# Observation:
According to the plot, year 1976-1978 was most profitable years and the profit was very low between the years 1960 and 1970.
